### String & functions



```python

# Python offers a variety of string manipulation functions that allow you to work with strings effectively. 
# Here's an explanation of some common string functions along with examples:

""" len(): Returns the length (the number of characters) of a string."""
text = "Hello, World!"
length = len(text)
print(length)  # Output: 13


""" str(): Converts other data types to strings."""
num = 42
text = str(num)
print(text)  # Output: "42"

"""concatenation (+): Combines two or more strings."""
str1 = "Hello"
str2 = "World"
result = str1 + ", " + str2
print(result)  # Output: "Hello, World"

""" .format(): Formats strings by replacing placeholders with values. """
name = "Alice"
age = 30
message = "My name is {} and I am {} years old.".format(name, age)
print(message)  # Output: "My name is Alice and I am 30 years old."

""" f-strings: A more concise way to format strings (Python 3.6+)."""
name = "Bob"
age = 25
message = f"My name is {name} and I am {age} years old."
print(message)  # Output: "My name is Bob and I am 25 years old."

""" .split(): Splits a string into a list of substrings based on a delimiter. """
sentence = "This is a sample sentence."
words = sentence.split(" ")
print(words)  # Output: ['This', 'is', 'a', 'sample', 'sentence.']

""" .join(): Joins a list of strings into a single string using a delimiter."""
words = ['This', 'is', 'a', 'list']
sentence = " ".join(words)
print(sentence)  # Output: "This is a list"

""" .strip(): Removes leading and trailing whitespace characters."""
text = "   Hello, World!   "
cleaned_text = text.strip()
print(cleaned_text)  # Output: "Hello, World!"

""" .upper() and .lower(): Converts a string to uppercase or lowercase."""
text = "Hello, World!"
upper_text = text.upper()
lower_text = text.lower()
print(upper_text)  # Output: "HELLO, WORLD!"
print(lower_text)  # Output: "hello, world!"

""" .replace(): Replaces all occurrences of a substring with another."""
text = "Hello, World!"
new_text = text.replace("World", "Python")
print(new_text)  # Output: "Hello, Python!"
```

```python
""" endswith() : Determines if string ends with given substring. """
""" startswith() : Determines if string starts with given substring. """
# endswith() and startswith()
my_string = "Hello, World!"
print(my_string.endswith("World!"))  # True
print(my_string.startswith("Hello"))  # True

""" isalnum() : Returns true if all characters are alphanumeric and false otherwise. """
""" isalpha() : Returns true if all characters are alphabetic and false otherwise. """
""" isdigit() : Returns true if string contains only digits. """
""" islower() : Returns true if string has all cased characters in lowercase and false otherwise. """
""" istitle() : Returns true if each word in string starts with uppercase character and others in lowercase. """
""" isupper() : Returns true if all characters in string are in uppercase and false otherwise. """

# isalnum(), isalpha(), isdigit(), islower(), isupper(), istitle()
print(my_string.isalnum())  # False (contains non-alphanumeric character ',')
print(my_string.isalpha())  # False (contains non-alphabetic character ',')
print(my_string.isdigit())  # False
print(my_string.islower())  # False
print(my_string.isupper())  # False
print(my_string.istitle())  # True

""" -------------------------------------------------------------------------------------------"""
""" find() : Determine if a substring occurs in given string. """ Returns index if found and -1 otherwise
""" index() : Same as find(), but raises an exception if substring not found. """
""" rfind() : Same as find(), but search string in reverse direction. """
""" rindex() : Same as index(), but search string in reverse direction. """

# find(), index(), rfind(), rindex()
print(my_string.find("World"))  # 7
print(my_string.find("Universe"))  # -1
print(my_string.index("World"))  # 7
# print(my_string.index("Universe"))  # ValueError: substring not found
print(my_string.rfind("o"))  # 8
print(my_string.rindex("o"))  # 8

""" #-------------------------------------------------------------------------------------------"""
""" center() :	Returns a string padded with given character so that original string is centered to a specified width. """
""" expandtabs() :	replaces tab escape sequence in string with multiple spaces.  default tab size is 8. """
""" ljust()	: string is padded with spaces on left and left-justified to a specified width. """
""" lstrip() : Removes leading whitespaces in string. """
""" rjust() : string is padded with spaces on right and right-justified to a specified width. """
""" rstrip() : Removes trailing whitespaces of string. """
""" split()	: Splits given string according to delimiter character or substring(default is space) and returns list of substrings. """
""" strip()	: Performs both lstrip() and rstrip() on string. """
""" ----------------------------------------------------------------------------------------------"""
# center(), expandtabs(), ljust(), rjust(), lstrip(), rstrip(), strip()
print(my_string.center(20, '-'))  # ---Hello, World!----
print('Hello\tWorld'.expandtabs())  # Hello   World
print(my_string.ljust(20, '-'))  # Hello, World!-------
print(my_string.rjust(20, '-'))  # -------Hello, World!
print('   Hello, World!  '.lstrip())  # 'Hello, World!  '
print('   Hello, World!  '.rstrip())  # '   Hello, World!'
print('   Hello, World!  '.strip())  # 'Hello, World!'

# split()
print(my_string.split(','))  # ['Hello', ' World!']
```

### Array/List Functions

```python

""" Creating Lists: """
my_list = []  # Empty list
my_list = [1, 2, 3]  # List with elements

""" Accessing Elements: """

my_list = [10, 20, 30, 40, 50]
print(my_list[0])  # Output: 10
print(my_list[-1])  # Output: 50

""" Modifying Lists: """

my_list = [1, 2, 3]
my_list.append(4)  # Adds 4 to the end
my_list.insert(1, 5)  # Inserts 5 at index 1
my_list.extend([6, 7])  # Adds [6, 7] to the end
my_list.remove(3)  # Removes the first 3
my_list.pop()  # Removes and returns the last element
my_list.pop(1)  # Removes and returns the element at index 1
my_list.clear()  # Removes all elements

""" Searching and Counting:"""

my_list = [10, 20, 30, 20, 40]
print(20 in my_list)  # Output: True
print(my_list.count(20))  # Output: 2

""" Sorting and Reversing: """

my_list = [40, 10, 30, 20]
my_list.sort()  # Sorts in ascending order in-place
print(my_list)  # Output: [10, 20, 30, 40]
my_list.sort(reverse=True)  # Sorts in descending order in-place
print(my_list)  # Output: [40, 30, 20, 10]
sorted_list = sorted(my_list)  # Creates a sorted copy
my_list.reverse()  # Reverses the list in-place
print(my_list)  # Output: [10, 20, 30, 40]


""" Slicing: """

my_list = [0, 1, 2, 3, 4, 5]
sublist = my_list[1:4]  # Extracts [1, 2, 3]
sublist_step = my_list[::2]  # Extracts [0, 2, 4]

""" Copying Lists: """

my_list = [1, 2, 3]
new_list = my_list.copy()  # Creates a shallow copy
new_list[0] = 99  # Doesn't affect the original list

""" Concatenating Lists: """

list1 = [1, 2]
list2 = [3, 4]
combined_list = list1 + list2  # Combines into [1, 2, 3, 4]
""" List Comprehensions: """

my_list = [1, 2, 3, 4, 5]
doubled_list = [x * 2 for x in my_list]  # [2, 4, 6, 8, 10]

""" Length and Range: """

my_list = [1, 2, 3, 4, 5]
length = len(my_list)  # Length: 5
my_range = list(range(1, 6))  # Generates [1, 2, 3, 4, 5]

""" List Aliasing
Consider the following code: """

lst1 = [11, 12, 13, 14, 15, 16, 17]
lst2 = lst1
lst1[-1] = 18
print(lst2)  # [11, 12, 13, 14, 15, 16, 18]

# After the second statement executes, lst1 and lst2 both refer to the same list. When two
# variables refer to the same objects, they are aliases. If that list is modified, both of lst1 and lst2
# will see the change.
# But be careful about:

lst1 = [11, 12, 13, 14, 15, 16, 17]
lst2 = lst1
lst1 = [5, 6]
print(lst2)  # [11, 12, 13, 14, 15, 16, 17]

# And also:

lst1 = [1, 2, 3]
lst2 = lst1[:]
lst2.remove(2)
print(lst1)  # [1,2,3]
```

### list Comprehension

```python
# List Comprehension
my_array = [11, 12, 13, 14, 15, 16, 17]
# Creating a new list using list comprehension
new_list = [x * 2 for x in my_array]
print(new_list)  # Output: [2, 4, 6, 8, 10]

# Creating a new list using list comprehension with condition
even_list = [x for x in my_array if x % 2 == 0]
print(even_list)  # Output: [2, 4]

# Creating a new list using list comprehension with condition and transformation
transformed_list = [x * 2 if x % 2 == 0 else x for x in my_array]
print(transformed_list)  # Output: [1, 4, 3, 8, 5]
```

### Tuple

```python
# Tuple Functions

# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
# or
my_tuple2 = tuple()
print(my_tuple, my_tuple2)

# Accessing elements in a tuple
print(my_tuple[0])  # Output: 1

# Modifying elements in a tuple (not possible)

# Finding the length of a tuple
length = len(my_tuple)
print(length)  # Output: 5

# Counting occurrences of an element in a tuple
count = my_tuple.count(3)
print(count)  # Output: 1

# Finding the index of an element in a tuple
index = my_tuple.index(4)
print(index)  # Output: 3

# Checking if an element exists in a tuple
if 2 in my_tuple:
    print("Element exists")
else:
    print("Element does not exist")

# Concatenating tuples
tuple1 = (1, 2)
tuple2 = (3, 4)
combined_tuple = tuple1 + tuple2
print(combined_tuple)  # Output: (1, 2, 3, 4)

# Slicing a tuple
subtuple = my_tuple[1:4]
print(subtuple)  # Output: (2, 3, 4)

# Converting a tuple to a list
my_list = list(my_tuple)
print(my_list)  # Output: [1, 2, 3, 4, 5]

# Converting a list to a tuple
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3, 4, 5)

# """ Tuple Packing and Unpacking: """
# You can pack multiple values into a tuple and then unpack them into variables. """

coordinates = (3, 4)
x, y = coordinates
print(x, y)  # 3 4

#Repeating Tuples:#
# Tuples can be repeated using the * operator.
my_tuple = (1, 2)
repeated_tuple = my_tuple * 3
print(repeated_tuple)  # (1, 2, 1, 2, 1, 2)
```

### Dictionary 

```python


# Dictionary Functions

# Creating a dictionary
my_dict = {"name": "Alice", "age": 30, "city": "New York"}

# Accessing values in a dictionary
print(my_dict["name"])  # Output: "Alice"

# Modifying values in a dictionary
my_dict["age"] = 31
print(my_dict)  # Output: {"name": "Alice", "age": 31, "city": "New York"}

# Adding new key-value pairs to a dictionary
my_dict["gender"] = "Female"
# Output: {"name": "Alice", "age": 31, "city": "New York", "gender": "Female"}
print(my_dict)

# Removing key-value pairs from a dictionary
del my_dict["city"]
print(my_dict)  # Output: {"name": "Alice", "age": 31, "gender": "Female"}

# Checking if a key exists in a dictionary
if "age" in my_dict:
    print("Key exists")
else:
    print("Key does not exist")

# Getting the number of key-value pairs in a dictionary
length = len(my_dict)
print(length)  # Output: 3

# Getting a list of all keys in a dictionary
keys = list(my_dict.keys())
print(keys)  # Output: ["name", "age", "gender"]

# Getting a list of all values in a dictionary
values = list(my_dict.values())
print(values)  # Output: ["Alice", 31, "Female"]

# Getting a list of all key-value pairs in a dictionary
items = list(my_dict.items())
print(items)  # Output: [("name", "Alice"), ("age", 31), ("gender", "Female")]

```

```python

### Dictionary class methods
# Dictionary creation and initialization
my_dict = dict()
print(my_dict)  # {}

# Adding key-value pairs
my_dict["name"] = "Alice"
my_dict["age"] = 30
my_dict["city"] = "New York"

# Shallow copy of dictionary
new_dict = my_dict.copy()
print(new_dict)  # {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Clearing all items from dictionary
my_dict.clear()
print(my_dict)  # {}

# Creating a dictionary using fromkeys()
keys = ['a', 'b', 'c']
default_value = 0
new_dict = dict.fromkeys(keys, default_value)
print(new_dict)  # {'a': 0, 'b': 0, 'c': 0}

# Using get() to retrieve a value associated with a key
print(new_dict.get('a'))  # 0

# Retrieving a list of key-value pairs using items()
print(new_dict.items())  # dict_items([('a', 0), ('b', 0), ('c', 0)])

# Retrieving a list of keys using keys()
print(new_dict.keys())  # dict_keys(['a', 'b', 'c'])

# Removing a specific key and returning the corresponding value using pop()
print(new_dict.pop('b'))  # 0

# Removing and returning some (key, value) pair using popitem()
print(new_dict.popitem())  # ('c', 0)

# Using setdefault() to get the value associated with a key or set it if the key is not present
print(new_dict.setdefault('d', 1))  # 1
print(new_dict)  # {'a': 0, 'd': 1}

# Updating dictionary from another dictionary using update()
update_dict = {'e': 2, 'f': 3}
new_dict.update(update_dict)
print(new_dict)  # {'a': 0, 'd': 1, 'e': 2, 'f': 3}

# Retrieving a list of values using values()
print(new_dict.values())  # dict_values([0, 1, 2, 3])
```

### set

```python

# Set Functions

# Creating a set
my_set = {1, 2, 3}

# Adding elements to a set
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

# Removing elements from a set
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4}

# Checking if an element exists in a set
if 2 in my_set:
    print("Element exists")
else:
    print("Element does not exist")

# Getting the number of elements in a set
length = len(my_set)
print(length)  # Output: 3

# Combining sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
combined_set = set1.union(set2)
print(combined_set)  # Output: {1, 2, 3, 4, 5}

# Finding the intersection of sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
intersection_set = set1.intersection(set2)
print(intersection_set)  # Output: {3}

# Finding the difference between sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
difference_set = set1.difference(set2)
print(difference_set)  # Output: {1, 2}

# Checking if a set is a subset of another set
set1 = {1, 2}
set2 = {1, 2, 3, 4, 5}
is_subset = set1.issubset(set2)
print(is_subset)  # Output: True

# Checking if a set is a superset of another set
set1 = {1, 2, 3, 4, 5}
set2 = {1, 2}
is_superset = set1.issuperset(set2)
print(is_superset)  # Output: True

# Checking if two sets have no elements in common
set1 = {1, 2, 3}
set2 = {4, 5, 6}
is_disjoint = set1.isdisjoint(set2)
print(is_disjoint)  # Output: True


s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
print("s1 before running difference_update: ", s1)
s1.difference_update(s2)
print("s1 after running difference_update: ", s1)  # {1, 2, 3}
-------------------------------------------------------------------------------------------
# Creating and initializing sets
my_set = {1, 2, 3, 4, 5}
other_set = {4, 5, 6, 7, 8}

# Adding an element to a set using add()
my_set.add(6)
print(my_set)  # {1, 2, 3, 4, 5, 6}

# Removing and returning an arbitrary set element using pop()
print(my_set.pop())

# Removing an element from a set using remove()
my_set.remove(2)
print(my_set)  # {1, 3, 4, 5, 6}

# Removing an element from a set using discard()
my_set.discard(3)
print(my_set)  # {1, 4, 5, 6}

# Updating a set with the union of itself and others using update()
my_set.update(other_set)
print(my_set)  # {1, 4, 5, 6, 7, 8}

# Performing set operations
print(my_set.difference(other_set))  # {1, 6}
my_set.difference_update(other_set)
print(my_set)  # {1, 6}

print(my_set.intersection(other_set))  # set()
my_set.intersection_update(other_set)
print(my_set)  # set()

print(my_set.symmetric_difference(other_set))  # {7, 8}
my_set.symmetric_difference_update(other_set)
print(my_set)  # {7, 8}

print(my_set.union(other_set))  # {4, 5, 6, 7, 8}
print(my_set)  # {7, 8}

# Checking set relations
print(my_set.isdisjoint(other_set))  # False
print(my_set.issubset(other_set))  # False
print(my_set.issuperset(other_set))  # False
```

### Loops

```python


# Loops:

""" For Loop: """

my_list = [1, 2, 3, 4, 5]
for num in my_list:
    print(num)

"""
Output:
1
2
3
4
5
"""

""" While Loop: """

i = 0
while i < 5:
    print(i)
    i += 1

"""
Output:
0
1
2
3
4
"""


""" Nested Loops: """

for i in range(3):
    for j in range(2):
        print(i, j)

"""
Output:
0 0
0 1
1 0
1 1
2 0
2 1
"""


""" Loop Control Statements: """

""" Break Statement: """

my_list = [1, 2, 3, 4, 5]
for num in my_list:
    if num == 3:
        break
    print(num)

"""
Output:
1
2
"""


""" Continue Statement: """

my_list = [1, 2, 3, 4, 5]
for num in my_list:
    if num == 3:
        continue
    print(num)

"""
Output:
1
2
4
5
"""


""" Pass Statement: """

my_list = [1, 2, 3, 4, 5]
for num in my_list:
    pass

"""
No output
"""
```

### Global and Local variables 

```python

# Global and Local variables

# Global variables are defined outside of any function and can be accessed and modified by any part of the code.
# Local variables are defined inside a function and can only be accessed and modified within that function.

# Example of Global variable:
global_var = 10


def my_function():
    print(global_var)  # Output: 10


my_function()

# Example of Local variable:


def my_function():
    local_var = 20
    print(local_var)  # Output: 20


my_function()

# Trying to access a local variable outside of its scope will result in an error:


def my_function():
    local_var = 20


my_function()
print(local_var)  # NameError: name 'local_var' is not defined
```

```python
x = 10  # This is a global variable


def my_function():
    global x  # Using the global keyword to indicate 'x' is a global variable
    x = 20  # Modifying the global variable 'x' within the function


my_function()
print(x)  # Output: 20 (the global variable 'x' has been modified)
```

# Unicode System in Python

```python

# Unicode is a standard for representing characters from different writing systems. In Python, strings are Unicode by default, which means they can represent characters from any language or script.

# Example:

unicode_string = "Hello, 你好, नमस्ते, مرحبا"
print(unicode_string)

"""
Output:
Hello, 你好, नमस्ते, مرحبا
"""

# Unicode Escape Sequence

# Unicode escape sequences allow you to represent Unicode characters using their hexadecimal code points. The syntax for a Unicode escape sequence is \u followed by the four-digit hexadecimal code point.

# Example:

unicode_string = "\u0048\u0065\u006C\u006C\u006F, \u4F60\u597D, \u0928\u092E\u0938\u094D\u0924\u0947, \u0645\u0631\u062D\u0628\u0627"
print(unicode_string)

"""
Output:
Hello, 你好, नमस्ते, مرحبا
"""

# Unicode Encoding and Decoding

# Encoding is the process of converting a Unicode string into a byte sequence, and decoding is the process of converting a byte sequence into a Unicode string.

# Example:

unicode_string = "Hello, 你好, नमस्ते, مرحبا"

# Encoding the Unicode string into bytes using UTF-8 encoding
encoded_bytes = unicode_string.encode("utf-8")
print(encoded_bytes)

"""
Output:
b'Hello, \xe4\xbd\xa0\xe5\xa5\xbd, \xe0\xa4\xa8\xe0\xa4\xae\xe0\xa4\xb8\xe0\xa5\x8d\xe0\xa4\xa4\xe0\xa5\x87, \xd9\x85\xd8\xb1\xd8\xad\xd8\xa8\xd8\xa7'
"""

# Decoding the bytes into a Unicode string using UTF-8 encoding
decoded_string = encoded_bytes.decode("utf-8")
print(decoded_string)

"""
Output:
Hello, 你好, नमस्ते, مرحبا
"""

# Note: The choice of encoding depends on the specific requirements of your application.
```

### Python - Operators

```python


""" Operators in Python """

# Arithmetic Operators

x = 10
y = 5

addition = x + y
print(addition)  # Output: 15

subtraction = x - y
print(subtraction)  # Output: 5

multiplication = x * y
print(multiplication)  # Output: 50

division = x / y
print(division)  # Output: 2.0

modulus = x % y
print(modulus)  # Output: 0

exponentiation = x ** y
print(exponentiation)  # Output: 100000

floor_division = x // y
print(floor_division)  # Output: 2


# Comparison Operators

x = 10
y = 5

greater_than = x > y
print(greater_than)  # Output: True

less_than = x < y
print(less_than)  # Output: False

equal_to = x == y
print(equal_to)  # Output: False

not_equal_to = x != y
print(not_equal_to)  # Output: True

greater_than_or_equal_to = x >= y
print(greater_than_or_equal_to)  # Output: True

less_than_or_equal_to = x <= y
print(less_than_or_equal_to)  # Output: False


# Assignment Operators

x = 10
y = 5

x += y
print(x)  # Output: 15

x -= y
print(x)  # Output: 10

x *= y
print(x)  # Output: 50

x /= y
print(x)  # Output: 10.0

x %= y
print(x)  # Output: 0.0

x **= y
print(x)  # Output: 0.0

x //= y
print(x)  # Output: 0.0


# Logical Operators

x = True
y = False

logical_and = x and y
print(logical_and)  # Output: False

logical_or = x or y
print(logical_or)  # Output: True

logical_not = not x
print(logical_not)  # Output: False


# Membership Operators

x = [1, 2, 3, 4, 5]

in_operator = 3 in x
print(in_operator)  # Output: True

not_in_operator = 6 not in x
print(not_in_operator)  # Output: True


# Identity Operators

x = 10
y = 10

is_operator = x is y
print(is_operator)  # Output: True

is_not_operator = x is not y
print(is_not_operator)  # Output: False


# Bitwise Operators

x = 10  # Binary: 1010
y = 5  # Binary: 0101

bitwise_and = x & y  # Binary: 0000
print(bitwise_and)  # Output: 0

bitwise_or = x | y  # Binary: 1111
print(bitwise_or)  # Output: 15

bitwise_xor = x ^ y  # Binary: 1111
print(bitwise_xor)  # Output: 15

bitwise_not = ~x  # Binary: 0101
print(bitwise_not)  # Output: -11

left_shift = x << 1  # Binary: 10100
print(left_shift)  # Output: 20

right_shift = x >> 1  # Binary: 0101
print(right_shift)  # Output: 5
```

### OOPS

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        pass


class Dog(Animal):
    def __init__(self, name, breed, toy):
        super().__init__(name, species="Dog")
        self.breed = breed
        self.toy = toy

    def make_sound(self):
        return "Woof!"


class Cat(Animal):
    def __init__(self, name, breed, favorite_food):
        super().__init__(name, species="Cat")
        self.breed = breed
        self.favorite_food = favorite_food

    def make_sound(self):
        return "Meow!"


my_dog = Dog("Fido", "Golden Retriever", "Ball")
my_cat = Cat("Whiskers", "Siamese", "Tuna")

print(my_dog.name)
print(my_dog.breed)
print(my_dog.toy)
print(my_dog.make_sound())

print(my_cat.name)
print(my_cat.breed)
print(my_cat.favorite_food)
print(my_cat.make_sound())
```

```python
## The itertools module
## The module of Python’s standard library consists of more efficient and fast iteration tools.

1. `count()`: Returns an iterator that generates evenly spaced values starting from the specified number.
2. `cycle()`: Returns each element from a given iterable and saves its copy to repeat the elements, forming an infinite loop.
3. `repeat()`: Returns the object argument repeatedly.
4. `accumulate()`: Applies a function to successive items in a list, storing the intermediate results.
5. `dropwhile()`: Returns an iterator by dropping elements from the iterable as long as the predicate function argument returns true.
6. `filterfalse()`: Returns an iterator by filtering out elements for which the predicate function results in False.
7. `islice()`: Builds an iterator by selecting certain elements from the iterable, similar to the Python slice function.
8. `product()`: Generates an iterator which is the Cartesian product of elements in the input sequences.
9. `permutations()`: Yields all possible permutations of elements in the input iterable.
10. `combinations()`: Yields all possible combinations of elements in the input iterable.

## These functions are particularly useful for efficient and optimized iteration operations in Python.
import itertools

# Example of count() function
for i in itertools.count(5, 2):
    if i > 20:
        break
    print(i, end=' ')  # Output: 5 7 9 11 13 15 17 19

# Example of cycle() function
count = 0
for item in itertools.cycle([1, 2, 3]):
    if count > 7:
        break
    print(item, end=' ')  # Output: 1 2 3 1 2 3 1 2

# Example of repeat() function
for i in itertools.repeat('Hello', 3):
    print(i, end=' ')  # Output: Hello Hello Hello

# Example of accumulate() function
data = [1, 2, 3, 4, 5]
result = itertools.accumulate(data, lambda x, y: x + y)
print(list(result))  # Output: [1, 3, 6, 10, 15]

# Example of dropwhile() function
data = [1, 3, 5, 2, 4, 6, 7]
result = itertools.dropwhile(lambda x: x < 5, data)
print(list(result))  # Output: [5, 2, 4, 6, 7]

# Example of filterfalse() function
data = [1, 0, 1, 0, 1, 0]
result = itertools.filterfalse(lambda x: x == 1, data)
print(list(result))  # Output: [0, 0, 0]

# Example of islice() function
data = 'ABCDEFG'
result = itertools.islice(data, 2, 5)
print(list(result))  # Output: ['C', 'D', 'E']

# Example of product() function
result = itertools.product([1, 2], ['a', 'b'])
print(list(result))  # Output: [(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')]

# Example of permutations() function
result = itertools.permutations([1, 2, 3], 2)
print(list(result))  # Output: [(1, 2), (1, 3), (2, 1), (2, 3), (3, 1), (3, 2)]

# Example of combinations() function
result = itertools.combinations([1, 2, 3, 4], 2)
print(list(result))  # Output: [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)]
```

### Class decorator
Just as function is a callable object, class instance also is a callable object. A callable object is an object which can be used and behaves like a function but might not be a function. To define class in a way that its instances should be callable objects, it should have  __call__ method. The instance is called "like a function", i.e. using brackets.

Class decorator contains a custom __class__() method. It returns a custom object that extends the functionality of what the original function does, in that case a class decorator should be used.

In following example add_decorator class has __call__ method. It modifies the arguments of decorated add() function. The add() function is designed to receive variable list of numbers. The decorator doubles those numbers before calling add() function.

```python
class add_decorator:

    def __init__(self, f):
        self.f = f

    def __call__(self, *args):
        print("original numbers:", args)
        # decorate add() before calling. doubles each number
        args = (2*i for i in args)
        self.f(*args)
        print('decorator terminated')
        # after f actions


@add_decorator
def add(*args):
    print("numbers received after decoration", args)
    s = 0
    for n in args:
        s = s+n
    print('addition of numbers', s)
```

```python
import math


class Triangle:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def area(self):
        s = (self.a+self.b+self.c)/2
        area = math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c))
        return area


if __name__ == '__main__':
    t1 = Triangle(4, 13, 15)
    print('area = ', t1.area())
```

## protected attributes
As mentioned earlier, Python doesn't impose restrictions on access of any class resource outside it. Python doesn't have keywords such as public, private or protected as in C++ or Java.   So in a way, all members in a Python class behave as if they have a public access.

```python
class MyClass:
    def __init__(self):
        self._protected_attribute = 42

    def get_protected_attribute(self):
        return self._protected_attribute

    def set_protected_attribute(self, value):
        if value >= 0:
            self._protected_attribute = value
        else:
            print("Value must be non-negative.")


# Create an instance of MyClass
obj = MyClass()

# Access the protected attribute using a getter method
print(obj.get_protected_attribute())  # Output: 42

# Try to access the protected attribute directly (not recommended)
print(obj._protected_attribute)  # Output: 42

# Modify the protected attribute using a setter method
obj.set_protected_attribute(100)

# Access the protected attribute using a getter method
print(obj.get_protected_attribute())  # Output: 100

# Try to set the protected attribute to a negative value (via setter)
obj.set_protected_attribute(-10)  # Output: Value must be non-negative.

# Access the protected attribute using a getter method
print(obj.get_protected_attribute())  # Output: 100 (unchanged)
```

## __repr__() and __str__() methods

## Operator overloading 
An operator is usually a symbol that is predefined to a certain operation. For instance arithmetic operators by default operate upon numeric operands. Hence only numeric objects are allowed to be used along with operators like +, -, *, / etc.

In Python, each operator actually invokes a corresponding method from object class that has two underscores in beginning and end. For example the '+' operator actually calls __add__() method as shown below:

```python
a = 10
b = 20
print(a+b)

# This is equivalent to

print(a.__add__(b))
```

```python
Following table shows magic methods to overload different operators.

Operator	Method
+	object.__add__(self, other)
-	object.__sub__(self, other)
*	object.__mul__(self, other)
// object.__floordiv__(self, other)
/	object.__div__(self, other)
%	object.__mod__(self, other)
** object.__pow__(self, other[, modulo])
<	object.__lt__(self, other)
<= object.__le__(self, other)
== object.__eq__(self, other)
!= object.__ne__(self, other)
>= object.__ge__(self, other)
```

# __add__() method
In following example, a class named MyTime is defined with two instance attributes hour and min. Addition of two MyTime objects is desired to be performed by overloading + operator.

To achieve this, __add__() magic method is overridden which performs addition of hour and min attributes of two objects. The __str__() method returns object’s string representation.

```python
class MyTime:

    def __init__(self, h=None, m=None):
        self.hour = h
        self.min = m

    def __add__(self, other):
        temptime = MyTime()
        temptime.hour = self.hour+other.hour
        temptime.min = self.min+other.min
        if temptime.min >= 60:
            temptime.hour = temptime.hour+1
            temptime.min = temptime.min-60
        return temptime

    def __sub__(self, other):
        temptime = MyTime()
        if self.min < other.min:
            self.min = self.min+60
            self.hour = self.hour-1
        temptime.hour = self.hour-other.hour
        temptime.min = self.min-other.min
        return temptime

    def __gt__(self, other):
        min1 = self.hour*60+self.min
        min2 = other.hour+other.min
        if min1 > min2:
            print('First time is greater than second')
        else:
            print('First time is not greater than second')

    def __str__(self):
        return str(self.hour)+" Hr."+str(self.min)+" min."
```

```python
a = MyTime(4, 30)
b = MyTime(2, 50)
print(a + b)
```

```python
a = MyTime(4, 30)
b = MyTime(2, 50)
print(a-b)
```

```python
normal = "computer"
print(normal)
raw = r"computer"
print(raw)
```

Both strings appear similar. The difference is evident when the string literal embeds escape characters ('\n', '\t' etc.)

Regular expressions use two types of characters in the matching pattern string: Meta characters are characters having a special meaning, similar to * in wild card. Literals are alphanumeric characters.

Following list of characters are called the metacharacters.

. ^ $ * + ? { } [ ] \ | ( )

```python
The square brackets[and ] are used for specifying a  set of characters that you wish to match. Characters can be listed individually, or a range of characters can be indicated by giving two characters and separating them by a '-'.

[abc]	Match any of the characters a, b, or c
[a-c]	Which uses a range to express the same set of characters.
[a-z]	Match only lowercase letters.
[0-9]	Match only digits.
'^'	Complements the character set in [].[ ^ 5] will match any character except '5'.
```

```python
'\'is an escaping metacharacter followed by various characters to signal various special sequences. If you need to match a [ or \, you can precede them with a backslash to remove their special meaning: \[ or \\.

Some of the special sequences beginning with '\' represent predefined sets of characters.

\d	Matches any decimal digit; this is equivalent to the class [0-9].
\D	Matches any non-digit character; this is equivalent to the class [ ^ 0-9].
\s	Matches any whitespace character; this is equivalent to the class [t\n\r\f\v].
\S	Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w	Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].
\W	Matches any non-alphanumeric character. equivalent to the class [ ^ a-zA-Z0-9_].
.	Matches with any single character except newline ‘\n’.
?	Match 0 or 1 occurrence of the pattern to its left
+ 1 or more occurrences of the pattern to its left
* 0 or more occurrences of the pattern to its left
\b	Boundary between word and non-word and /B is opposite of / b
[..]	Matches any single character in a square bracket and [ ^ ..] matches any single character not in square bracket
\	It is used for special meaning characters like \. to match a period or \+ for plus sign.
{n, m}	Matches at least n and at most m occurrences of preceding
a | b	Matches either a or b
```

The re module has following functions:

## re.match():
This method finds match for the pattern if it occurs at start of the string.

re.match(pattern, string)

```python
import re
string = "Simple is better than complex."
obj = re.match(r"Simple", string)
print(obj)
print(obj.start())
print(obj.end())
```

The match object's start() method returns the starting position of pattern in the string, and end() returns the endpoint.

If the pattern is not found, the match object is None.

This function searches for first occurrence of RE pattern within string from any position of the string but it only returns the first occurrence of the search pattern.

```python
import re
string = "Simple is better than complex."
obj = re.search(r"is", string)
print(obj.start())

print(obj.end())
```

This function returns a string by replacing a certain pattern by its substitute string. Usage of this function is :

## re.compile():
This function compiles a regular expression pattern into a regular expression object. This is useful when you need to use an expression several times.

```python
import re

string = 'Simple is better than complex. Complex is better than complicated.'
pattern = re.compile(r'is')

# Using match()
obj = pattern.match(string)
print(f"Match found at start: {obj.start() if obj else None}")
print(f"Match found at end: {obj.end() if obj else None}")

# Using search()
obj = pattern.search(string)
print(f"Search found at start: {obj.start() if obj else None}")
print(f"Search found at end: {obj.end() if obj else None}")

# Using findall()
obj = pattern.findall(string)
print(f"Findall results: {obj}")

# Using sub()
obj = pattern.sub(r'was', string)
print(f"String after substitution: {obj}")
```

## Python - Send Email
A mail server is an application that handles and delivers e-mail over the Internet. Mail servers are of two main categories: outgoing mail servers and incoming mail servers.

Outgoing mail servers are known as SMTP, or Simple Mail Transfer Protocol, servers. Simple Mail Transfer Protocol (SMTP) is an Internet standard for email transmission.

Incoming mail servers come in two main varieties. POP3, or Post Office Protocol and IMAP, or Internet Message Access Protocol.

Python's standard library has 'smtplib' module which defines an SMTP client session object that can be used to send mail via Python program.

### smptlib.SMTP() function:
This function returns object of SMTPclass. It encapsulates and manages a connection to an SMTP or ESMTP server. The function takes following arguments:



```python
host	Is the name of the remote host to which to connect.
port	Specifies the port to which to connect. By default, smtplib.SMTP_PORT is used.
local_hostname	Used as the FQDN of the local host in the HELO/EHLO command.
source_address	A 2-tuple(host, port) object for the socket to bind


It has following methods that support SMTP operations:

connect(host, port, source_address)	This method establishes connection to a host on a given port.
login(user, password)	Log in on an SMTP server that requires authentication.
quit()	terminate the SMTP session.
data(msg)	sends message data to server.
docmd(cmd, args)	send a command, and return its response code.
ehlo(name)	Hostname to identify itself
starttls()	puts the connection to the SMTP server into TLS mode.
getreply()	get a reply from the server consisting of server response code.
putcmd(cmd, args)	sends a command to the server.
send_message(msg, from_addr, to_addrs)	converts message to a bytestring and passes it to sendmail.
sendmail(self, from_addr, to_addrs, msg)
This command performs an entire mail transaction.
The arguments are:

from_addr	The address sending this mail.
to_addrs	A list of addresses to send this mail to.
msg	The message to send.
```
